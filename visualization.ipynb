```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


def gdp_plot():

    yearly_gdp = pd.read_csv("./src/yearly_gdp.csv")

    sns.set_style("whitegrid")

    sns.lineplot(yearly_gdp, x= "Year",  y="GDPES", label='Nominal GDP' )
    sns.lineplot(yearly_gdp, x = "Year", y="RGDPES", label='Real GDP')

    plt.title('Real vs Nominal GDP Over Time')
    plt.xlabel('Year')
    plt.ylabel('GDP (in billions)')

    plt.legend()
    plt.show()

```

```python
gdp_plot()
```

```python

def real_vs_nom_plot():

    yearly_wages = pd.read_csv("./src/yearly_wages.csv")

    sns.lineplot(yearly_wages, x= "Year",  y="Salary", label='Nominal Salary', linestyle='--')
    sns.lineplot(yearly_wages, x = "Year", y="Real Salary", label= "Real Salary (adj. CPI)", linestyle='-')

    sns.scatterplot(data=yearly_wages, x="Year", y="Salary", color='blue')
    sns.scatterplot(data=yearly_wages, x="Year", y="Real Salary", color='orange')

    plt.title('Nominal vs Real Salary (mean, adjusted for CPI)')
    plt.xlabel('Year')
    plt.ylabel('Salary (in thousands, EUR)')

    plt.yticks(range(0, 50, 12))

    plt.legend()

    plt.show()
```

```python
real_vs_nom_plot()
```

```python
def ibex_plot():

    yearly_close = pd.read_csv("./src/yearly_close.csv")
    sns.lineplot(yearly_close, x = "Year", y = "Close")

    plt.title('IBEX 35 performance (EOY close)')
    plt.xlabel('Year')
    plt.ylabel('Value (in EUR)')

    plt.legend()
    plt.show()
```

```python
ibex_plot()
```
